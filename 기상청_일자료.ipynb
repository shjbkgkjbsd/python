##Surface Temperature

```python
import matplotlib.pyplot as plt
df['tm'] = pd.to_datetime(df['tm'])
df['ts'] = pd.to_numeric(df['ts'], downcast='float')

fig = plt.figure(figsize=(14,6)) #사가형 크기
ax = plt.subplot(1,1,1)
ax.plot(df['tm'], df['ts'], color='r', lw=1.5, alpha=0.7)

ax.set_xlabel('Date날자', fontweight='bold', fontsize=12)
ax.set_ylabel('Surface Temperature', fontweight='bold',fontsize=12)
ax.set_title('OPEN API ASOS 온도 데이터', fontweight='bold',fontsize=12)
```

```python
import matplotlib.pyplot as plt
df['tm'] = pd.to_datetime(df['tm'])
df['ta'] = pd.to_numeric(df['ta'], downcast='float')

fig = plt.figure(figsize=(14,6)) #사가형 크기
ax = plt.subplot(1,1,1)
ax.plot(df['tm'], df['ta'], color='blue', lw=1.5, alpha=0.7)

ax.set_xlabel('Date날자', fontweight='bold', fontsize=12)
ax.set_ylabel('Surface Temperature', fontweight='bold',fontsize=12)
ax.set_title('OPEN API ASOS 온도 데이터', fontweight='bold',fontsize=12)
```

비교 출력(Surface & Air Temp)

```python
fig = plt.figure(figsize=(14,6)) 
ax = fig.add_subplot(1,1,1)
ax.plot(df['tm'], df['ts'], color='blue', lw=1.5, alpha=0.7, label='표면온도')
ax.plot(df['tm'], df['ta'], color='red', lw=1.5, alpha=0.7, label='Air')
ax.legend()
plt.show()
```
